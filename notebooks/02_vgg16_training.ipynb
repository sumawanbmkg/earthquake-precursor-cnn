```python
# SEL 1: IMPOR LIBRARY
# ==============================================================================

print("Menjalankan Sel 1: Impor Library...")
import tensorflow as tf
from tensorflow import keras
# Mengimpor lapisan secara langsung agar bisa menggunakan 'layers.Flatten()' dll.
from tensorflow.keras import layers 
from tensorflow.keras import backend as K
from tensorflow.keras.layers import Dense, Activation, Dropout, Conv2D, MaxPooling2D, BatchNormalization, Flatten, Input
from tensorflow.keras.optimizers import Adam
# categorical_crossentropy masih relevan jika output model Dense(NUM_CLASSES, activation='softmax')
from tensorflow.keras.metrics import categorical_crossentropy 
from tensorflow.keras import regularizers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Model, load_model, Sequential
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import os
import seaborn as sns
sns.set_style('darkgrid')
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, roc_auc_score
from IPython.display import display, HTML
from sklearn.utils import class_weight # Untuk menghitung class weights
print("Sel 1 Selesai.")

# ==============================================================================
# SEL 2: MEMBUAT DATAFRAME DARI JALUR FILE DAN LABEL
# ==============================================================================

print("\nMenjalankan Sel 2: Membuat DataFrame dari Jalur File dan Label...")

# --- KONFIGURASI PATH DATASET ---
# Path utama ke direktori yang berisi 'dataset_gambar_3komponen'
base_data_dir = r'/home/kmtr/Earthquake_Prediction/'
# Subfolder spesifik yang berisi kelas-kelas Anda
data_component_folder = 'dataset_gambar_3komponen'

# Path lengkap ke folder dataset yang berisi subfolder kelas
sdir = os.path.join(base_data_dir, data_component_folder)

# --- Kelas yang Diharapkan ---
class_names = ['normal', 'precursor'] # Nama folder kelas Anda
label_map = {name: i for i, name in enumerate(class_names)} # {'normal': 0, 'precursor': 1}
NUM_CLASSES = len(class_names) # Harusnya 2

print(f"Direktori dataset utama yang dicari: '{sdir}'")
print(f"Kelas yang diharapkan: {class_names}")

if not os.path.exists(sdir):
    raise FileNotFoundError(f"Direktori dataset tidak ditemukan di lokasi yang diharapkan: '{sdir}'")

filepaths = []
labels = []

# Memastikan semua folder kelas ada dan tidak kosong
found_all_classes = True
for class_name_expected in class_names:
    class_folder = os.path.join(sdir, class_name_expected)
    print(f"Mengecek subfolder kelas: {class_folder}")
    
    if not os.path.exists(class_folder):
        print(f"Peringatan: Subfolder '{class_name_expected}' tidak ditemukan di '{sdir}'.")
        found_all_classes = False
    else:
        # Filter hanya file gambar yang valid
        files_in_folder = [f for f in os.listdir(class_folder) if f.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp'))]
        if not files_in_folder:
            print(f"Peringatan: Subfolder '{class_name_expected}' kosong atau tidak mengandung file gambar yang didukung.")
            found_all_classes = False
        else:
            print(f"Subfolder '{class_name_expected}' ditemukan dengan {len(files_in_folder)} gambar.")

if not found_all_classes:
    raise FileNotFoundError(
        f"Salah satu atau lebih folder kelas ('normal', 'precursor') tidak ditemukan "
        f"atau kosong di '{sdir}'. Pastikan struktur folder dan nama kelas sudah benar "
        f"dan ada gambar di dalamnya (di '{sdir}/normal/' dan '{sdir}/precursor/')."
    )

print("\nMemulai pemuatan path gambar dan label...")
for klass in class_names:
    classpath = os.path.join(sdir, klass)
    for f in os.listdir(classpath):
        fpath = os.path.join(classpath, f)
        if os.path.isfile(fpath) and f.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp')):
            filepaths.append(fpath)
            labels.append(klass)

if not filepaths:
    raise ValueError("Tidak ada gambar yang dimuat. Pastikan path dan file gambar Anda benar.")

Fseries = pd.Series(filepaths, name='filepaths')
Lseries = pd.Series(labels, name='labels')
df = pd.concat([Fseries, Lseries], axis=1)

print("\n5 baris pertama dari DataFrame:")
print(df.head())
print("\nDistribusi kelas dalam DataFrame:")
print(df['labels'].value_counts())
print(f"Jumlah total gambar yang ditemukan: {len(df)}")
print("Sel 2 Selesai.")

# ==============================================================================
# SEL 3: MEMBAGI DATAFRAME MENJADI TRAIN, TEST, DAN VALIDATION & HITUNG CLASS WEIGHTS
# ==============================================================================

print("\nMenjalankan Sel 3: Membagi Data & Menghitung Class Weights...")

train_split = .8
test_split = .1
dummy_split = test_split / (1 - train_split) # Ini adalah test_size untuk split kedua

# Menggunakan stratify untuk menjaga proporsi kelas yang sama di setiap set
train_df, dummy_df = train_test_split(df, train_size=train_split, shuffle=True, random_state=123, stratify=df['labels'])
test_df, valid_df = train_test_split(dummy_df, train_size=dummy_split, shuffle=True, random_state=123, stratify=dummy_df['labels'])

print(f"Panjang train_df: {len(train_df)}")
print(f"Panjang valid_df: {len(valid_df)}")
print(f"Panjang test_df: {len(test_df)}")

print("\nDistribusi kelas di train_df:")
print(train_df['labels'].value_counts())
print("\nDistribusi kelas di valid_df:")
print(valid_df['labels'].value_counts())
print("\nDistribusi kelas di test_df:")
print(test_df['labels'].value_counts())

# Hitung class weights untuk menangani ketidakseimbangan kelas (jika ada)
# Ini penting karena ImageDataGenerator tidak memiliki parameter 'sample_weight' langsung
print("\nMenghitung bobot kelas untuk training...")
class_weights_array = class_weight.compute_class_weight(
    class_weight='balanced',
    classes=np.unique(train_df['labels']),
    y=train_df['labels']
)
# Ubah array bobot menjadi dictionary dengan label angka yang diharapkan Keras
# Misalnya: {'normal': 0, 'precursor': 1} maka class_weights[0] untuk 'normal', class_weights[1] untuk 'precursor'
computed_class_weights = {label_map[label]: weight for label, weight in zip(np.unique(train_df['labels']), class_weights_array)}

# --- Opsional: Tingkatkan bobot kelas 'precursor' secara manual jika recall adalah prioritas utama ---
# Anda bisa menyesuaikan faktor pengali ini untuk lebih menekankan deteksi 'precursor'
precursor_label_idx = label_map['precursor'] # Biasanya 1 jika 'normal' = 0
if precursor_label_idx in computed_class_weights:
    computed_class_weights[precursor_label_idx] *= 1.5 # Contoh: Tingkatkan bobot 1.5 kali
    print(f"Bobot kelas 'precursor' (label {precursor_label_idx}) ditingkatkan menjadi: {computed_class_weights[precursor_label_idx]:.4f}")

print(f"Bobot Kelas Akhir yang Digunakan: {computed_class_weights}")

print("Sel 3 Selesai.")

# ==============================================================================
# SEL 4: MEMBUAT IMAGE DATA GENERATORS
# ==============================================================================

print("\nMenjalankan Sel 4: Membuat Image Data Generators...")

height = 128
width = 128
channels = 3
batch_size = 32 # Ukuran batch yang umum

img_shape = (height, width, channels)
img_size = (height, width)

# Menghitung test_batch_size dan test_steps
length = len(test_df)
test_batch_size = batch_size
if test_batch_size > length: # Hindari batch size lebih besar dari data
    test_batch_size = length
test_steps = int(np.ceil(length / test_batch_size))

print(f'Ukuran gambar: {img_size}')
print(f'Batch size training/validasi: {batch_size}')
print(f'Test batch size: {test_batch_size}, Test steps: {test_steps}')

# Data Augmentation untuk Training
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

# Hanya rescaling untuk Validasi dan Test (PENTING: Jangan augmentasi data validasi/test)
valid_test_datagen = ImageDataGenerator(rescale=1./255)

train_gen = train_datagen.flow_from_dataframe(
    train_df, x_col='filepaths', y_col='labels',
    target_size=img_size, class_mode='categorical',
    color_mode='rgb', shuffle=True, batch_size=batch_size,
    # Pastikan class_indices sesuai dengan label_map Anda jika diperlukan penyesuaian
    # class_indices=label_map # Ini akan memastikan urutan kelas konsisten
)

valid_gen = valid_test_datagen.flow_from_dataframe(
    valid_df, x_col='filepaths', y_col='labels',
    target_size=img_size, class_mode='categorical',
    color_mode='rgb', shuffle=True, batch_size=batch_size,
    # class_indices=label_map
)

test_gen = valid_test_datagen.flow_from_dataframe(
    test_df, x_col='filepaths', y_col='labels',
    target_size=img_size, class_mode='categorical',
    color_mode='rgb', shuffle=False, batch_size=test_batch_size, # Shuffle=False untuk evaluasi konsisten
    # class_indices=label_map
)

classes = list(train_gen.class_indices.keys())
print(f"\nKelas yang ditemukan oleh generator: {classes}")
print(f"Jumlah kelas: {NUM_CLASSES}")

# Memastikan urutan kelas di generator sesuai dengan label_map jika ada perbedaan
# print(f"Class indices dari train_gen: {train_gen.class_indices}")
# Jika Anda ingin memastikan 'normal' selalu 0 dan 'precursor' selalu 1,
# Anda bisa memverifikasi train_gen.class_indices atau mengaturnya secara eksplisit saat membuat generator.
# Biasanya, `flow_from_dataframe` akan menetapkannya secara alfabetis jika tidak ditentukan.

print("Sel 4 Selesai.")

# ==============================================================================
# SEL 5: FUNGSI UNTUK MENAMPILKAN CONTOH GAMBAR
# ==============================================================================

print("\nMenjalankan Sel 5: Fungsi untuk Menampilkan Contoh Gambar...")

def show_image_samples(gen):
    # Mengambil class_indices dari generator yang digunakan untuk test
    class_dict = gen.class_indices
    classes = list(class_dict.keys())
    
    images, labels = next(gen) # Mendapatkan satu batch sampel dari generator
    
    plt.figure(figsize=(20, 20))
    length = len(labels)
    r = min(length, 25) # Tampilkan maksimal 25 gambar
    
    for i in range(r):
        plt.subplot(5, 5, i + 1)
        image = images[i]
        plt.imshow(image)
        
        # labels adalah one-hot encoded, jadi ambil indeks kelas
        index = np.argmax(labels[i])
        class_name = classes[index]
        plt.title(class_name, color='blue', fontsize=16)
        plt.axis('off')
    plt.show()

print("Sel 5 Selesai.")

# ==============================================================================
# SEL 6: MENAMPILKAN CONTOH GAMBAR DARI TRAINING SET
# ==============================================================================

print("\nMenjalankan Sel 6: Menampilkan Contoh Gambar dari Training Set...")
print("Ini adalah contoh gambar yang akan digunakan untuk training (sudah direscale/augmentasi jika berlaku).")
show_image_samples(train_gen)
print("Sel 6 Selesai.")
print(f"\n{'='*25} DATA PREPARATION SELESAI {'='*25}")


# ==============================================================================
# SEL 7: MEMBANGUN MODEL VGG16 (TRANSFER LEARNING)
# ==============================================================================

print(f"\n{'='*25} MEMBANGUN MODEL {'='*25}")
print("\nMenjalankan Sel 7: Membangun Model VGG16...")

# Memuat base model VGG16 dengan pre-trained weights dari ImageNet
# include_top=False: tidak menyertakan fully connected layer di bagian atas (akhir) model
# input_tensor: menentukan shape input gambar
base_model = tf.keras.applications.VGG16(
    include_top=False,
    weights="imagenet",
    input_tensor=Input(shape=img_shape) # Menggunakan img_shape (128,128,3)
)

# Membekukan lapisan-lapisan dari base model
# Ini penting dalam transfer learning agar fitur yang sudah dipelajari tidak rusak
base_model.trainable = False

print("\nRingkasan Base Model (VGG16 Feature Extractor):")
base_model.summary()
print("Base Model VGG16 telah dimuat dan lapisan-lapisan dibekukan.")

# Membangun model Sequential: base_model + lapisan klasifikasi kustom
model_name = 'VGG16_Earthquake_Predictor'
print(f"\nBuilding model '{model_name}' dengan base model VGG16.")

model = tf.keras.Sequential([
    base_model, # VGG16 sebagai feature extractor
    layers.Flatten(), # Meratakan output dari base_model menjadi 1D
    layers.Dense(256, activation='relu'), # Lapisan Dense kustom pertama
    layers.Dropout(0.5), # Dropout untuk mencegah overfitting
    layers.Dense(NUM_CLASSES, activation='softmax') # Lapisan output untuk 2 kelas (normal/precursor)
])

# Mengkompilasi model
# Menggunakan 'Adam' optimizer
# Menggunakan 'categorical_crossentropy' karena output adalah one-hot encoded
# Menambahkan metrik 'precision', 'recall', dan 'auc' untuk evaluasi yang lebih baik
model.compile(
    optimizer=Adam(learning_rate=.0001), # Learning rate yang lebih kecil untuk transfer learning
    loss='categorical_crossentropy',
    metrics=['accuracy',
             tf.keras.metrics.Precision(name='precision'),
             tf.keras.metrics.Recall(name='recall'),
             tf.keras.metrics.AUC(name='auc')]
)

print("\nRingkasan Model Akhir:")
model.summary()
print("Sel 7 Selesai.")


# ==============================================================================
# SEL 8: PELATIHAN MODEL
# ==============================================================================

print("\nMenjalankan Sel 8: Pelatihan Model...")

epochs = 30 # Jumlah epoch, Anda bisa menyesuaikan ini
# Callback EarlyStopping untuk mencegah overfitting dan menghemat waktu
early_stopping = EarlyStopping(
    monitor='val_loss', # Memantau loss pada data validasi
    patience=7,         # Berhenti jika val_loss tidak membaik selama 7 epoch
    restore_best_weights=True, # Mengembalikan bobot model dari epoch terbaik
    verbose=1
)
# Callback ModelCheckpoint untuk menyimpan model terbaik berdasarkan val_loss
model_checkpoint = ModelCheckpoint(
    filepath='best_vgg16_model.keras', # Simpan dalam format .keras
    monitor='val_loss',
    save_best_only=True,
    verbose=1
)

print(f"Model akan dilatih selama maksimal {epochs} epoch dengan EarlyStopping (patience={early_stopping.patience}).")
print(f"Model terbaik akan disimpan berdasarkan 'val_loss'.")

history = model.fit(
    x=train_gen,
    epochs=epochs,
    validation_data=valid_gen,
    class_weight=computed_class_weights, # Menggunakan bobot kelas yang sudah dihitung
    callbacks=[early_stopping, model_checkpoint]
)

print("Sel 8 Selesai.")
print(f"\n{'='*25} PELATIHAN MODEL SELESAI {'='*25}")

# ==============================================================================
# SEL 9: FUNGSI UNTUK MENAMPILKAN GRAFIK TRAINING
# ==============================================================================

print(f"\n{'='*25} ANALISIS HASIL PELATIHAN {'='*25}")
print("\nMenjalankan Sel 9: Fungsi untuk Menampilkan Grafik Training...")

def tr_plot(tr_data, start_epoch):
    # Plot the training and validation data
    acc = tr_data.history.get('accuracy')
    val_acc = tr_data.history.get('val_accuracy')
    loss = tr_data.history.get('loss')
    val_loss = tr_data.history.get('val_loss')
    precision = tr_data.history.get('precision')
    val_precision = tr_data.history.get('val_precision')
    recall = tr_data.history.get('recall')
    val_recall = tr_data.history.get('val_recall')
    auc = tr_data.history.get('auc')
    val_auc = tr_data.history.get('val_auc')

    epochs = range(start_epoch + 1, len(acc) + start_epoch + 1) # Menyesuaikan epoch
    
    # Mencari epoch terbaik
    index_loss = np.argmin(val_loss)
    val_lowest_loss = val_loss[index_loss]
    index_acc = np.argmax(val_acc)
    acc_highest = val_acc[index_acc]

    plt.style.use('fivethirtyeight')
    fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(20, 15)) # 2x2 grid for more metrics

    # Plot Loss
    axes[0, 0].plot(epochs, loss, 'r', label='Training Loss')
    axes[0, 0].plot(epochs, val_loss, 'g', label='Validation Loss')
    axes[0, 0].scatter(index_loss + start_epoch + 1, val_lowest_loss, s=150, c='blue', label=f'Best Epoch (Loss) = {index_loss + start_epoch + 1}')
    axes[0, 0].set_title('Training and Validation Loss', fontsize=16)
    axes[0, 0].set_xlabel('Epochs')
    axes[0, 0].set_ylabel('Loss')
    axes[0, 0].legend()
    axes[0, 0].grid(True)

    # Plot Accuracy
    axes[0, 1].plot(epochs, acc, 'r', label='Training Accuracy')
    axes[0, 1].plot(epochs, val_acc, 'g', label='Validation Accuracy')
    axes[0, 1].scatter(index_acc + start_epoch + 1, acc_highest, s=150, c='blue', label=f'Best Epoch (Accuracy) = {index_acc + start_epoch + 1}')
    axes[0, 1].set_title('Training and Validation Accuracy', fontsize=16)
    axes[0, 1].set_xlabel('Epochs')
    axes[0, 1].set_ylabel('Accuracy')
    axes[0, 1].legend()
    axes[0, 1].grid(True)

    # Plot Precision (jika ada)
    if precision is not None and val_precision is not None:
        axes[1, 0].plot(epochs, precision, 'purple', label='Training Precision')
        axes[1, 0].plot(epochs, val_precision, 'orange', label='Validation Precision')
        axes[1, 0].set_title('Training and Validation Precision', fontsize=16)
        axes[1, 0].set_xlabel('Epochs')
        axes[1, 0].set_ylabel('Precision')
        axes[1, 0].legend()
        axes[1, 0].grid(True)
    else:
        axes[1, 0].text(0.5, 0.5, 'Precision data not available', horizontalalignment='center', verticalalignment='center', transform=axes[1, 0].transAxes)

    # Plot Recall (jika ada)
    if recall is not None and val_recall is not None:
        axes[1, 1].plot(epochs, recall, 'brown', label='Training Recall')
        axes[1, 1].plot(epochs, val_recall, 'pink', label='Validation Recall')
        axes[1, 1].set_title('Training and Validation Recall', fontsize=16)
        axes[1, 1].set_xlabel('Epochs')
        axes[1, 1].set_ylabel('Recall')
        axes[1, 1].legend()
        axes[1, 1].grid(True)
    else:
        axes[1, 1].text(0.5, 0.5, 'Recall data not available', horizontalalignment='center', verticalalignment='center', transform=axes[1, 1].transAxes)

    plt.tight_layout()
    plt.show()

print("Sel 9 Selesai.")

# ==============================================================================
# SEL 10: MENAMPILKAN GRAFIK AKURASI DAN LOSS
# ==============================================================================

print("\nMenjalankan Sel 10: Menampilkan Grafik Akurasi dan Loss (serta Precision, Recall)...")
tr_plot(history, 0)
print("Sel 10 Selesai.")

# ==============================================================================
# SEL 11: FUNGSI UNTUK PRINT TEXT DALAM WARNA (HELPER)
# ==============================================================================

print("\nMenjalankan Sel 11: Fungsi untuk Print Text dalam Warna...")

def print_in_color(txt_msg, fore_tupple, back_tupple):
    #prints the text_msg in the foreground color specified by fore_tupple with the background specified by back_tupple
    #text_msg is the text, fore_tupple is foreground color tupple (r,g,b), back_tupple is background tupple (r,g,b)
    rf, gf, bf = fore_tupple
    rb, gb, bb = back_tupple
    msg = '{0}' + txt_msg
    mat = '\33[38;2;' + str(rf) + ';' + str(gf) + ';' + str(bf) + ';48;2;' + str(rb) + ';' + str(gb) + ';' + str(bb) + 'm'
    print(msg.format(mat), flush=True)
    print('\33[0m', flush=True) # returns default print color to back to black
    return

print("Sel 11 Selesai.")

# ==============================================================================
# SEL 12: EVALUASI MODEL PADA DATA TEST DAN SIMPAN MODEL TERBAIK
# ==============================================================================

print("\nMenjalankan Sel 12: Evaluasi Model pada Data Test dan Simpan Model Terbaik...")

# Muat kembali model terbaik dari checkpoint
try:
    best_model = tf.keras.models.load_model('best_vgg16_model.keras')
    print("Model terbaik berhasil dimuat dari 'best_vgg16_model.keras'.")
except Exception as e:
    print(f"Gagal memuat model terbaik: {e}. Menggunakan model yang terakhir dilatih.")
    best_model = model # Fallback ke model yang terakhir dilatih

results = best_model.evaluate(test_gen, steps=test_steps, return_dict=True, verbose=1)

print("\n--- Hasil Evaluasi pada Test Set ---")
for name, value in results.items():
    print(f"{name}: {value:.4f}")

acc_test = results['accuracy'] * 100
msg = f'Akurasi pada test set adalah {acc_test:5.2f} %'
print_in_color(msg, (0,255,0),(55,65,80))

# Jika model sudah disimpan oleh ModelCheckpoint, kita tidak perlu menyimpannya lagi di sini
# Namun, jika Anda ingin menyimpan dengan nama spesifik berdasarkan akurasi, bisa dilakukan
# save_dir = r'./' # Direktori penyimpanan
# save_id = str(model_name + '-' + str(acc_test)[:str(acc_test).rfind('.')+3] + '.keras')
# save_loc = os.path.join(save_dir, save_id)
# best_model.save(save_loc)
# print(f"Model terbaik juga disimpan dengan nama: {save_loc}")


print("Sel 12 Selesai.")

# ==============================================================================
# SEL 13: MEMBUAT PREDIKSI DAN MENAMPILKAN CONFUSION MATRIX SERTA CLASSIFICATION REPORT
# ==============================================================================

print("\nMenjalankan Sel 13: Membuat Prediksi dan Menampilkan Confusion Matrix & Classification Report...")

# Reset test_gen untuk memastikan urutan gambar konsisten
test_gen.reset()
preds_probs = best_model.predict(test_gen, verbose=1)

# Mengambil label true dari test_gen
y_true = test_gen.classes
# Mengambil nama kelas dari test_gen
class_labels = list(test_gen.class_indices.keys())
# Mengambil prediksi kelas (dari one-hot encoding ke label angka)
y_pred_labels = np.argmax(preds_probs, axis=1)

print("\n a. Tabel Laporan Klasifikasi (Precision, Recall, F1-Score):")
print(classification_report(y_true, y_pred_labels, target_names=class_labels, zero_division=0))

print("\n b. Grafik Confusion Matrix:")
cm = confusion_matrix(y_true, y_pred_labels)
cm_normalized = np.zeros_like(cm, dtype=float)
with np.errstate(divide='ignore', invalid='ignore'):
    cm_normalized = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
    cm_normalized[np.isnan(cm_normalized)] = 0 # Menangani kasus pembagian dengan nol jika ada kelas tanpa sampel
    
plt.figure(figsize=(8, 6))
sns.heatmap(cm_normalized, annot=True, fmt='.2%', cmap='Blues',
            xticklabels=class_labels,
            yticklabels=class_labels)
plt.title('Confusion Matrix (Ternormalisasi)', fontsize=16)
plt.ylabel('Kelas Aktual (Sebenarnya)')
plt.xlabel('Kelas Prediksi (Tebakan Model)')
plt.show()

print("\n c. Grafik Kurva ROC dan Nilai AUC:")
# Untuk binary classification, kita bisa langsung pakai probabilitas kelas positif (biasanya indeks 1)
# Pastikan 'precursor' adalah kelas 1
if 'precursor' in label_map and label_map['precursor'] == 1:
    y_score = preds_probs[:, 1] # Probabilitas kelas 'precursor'
else:
    # Jika 'precursor' adalah kelas 0 atau tidak pasti, kita harus mengidentifikasinya
    # Asumsi: kelas positif adalah kelas dengan indeks 1 di `class_labels`
    print("Peringatan: Kelas 'precursor' tidak ditemukan sebagai indeks 1. Menggunakan probabilitas kelas 1 secara default.")
    y_score = preds_probs[:, 1] # Fallback, mungkin perlu penyesuaian jika 'normal' adalah 1

auc_score = roc_auc_score(y_true, y_score)
fpr, tpr, thresholds = roc_curve(y_true, y_score)

plt.figure(figsize=(8, 8))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'Kurva ROC (AUC = {auc_score:.3f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Tebakan Acak (AUC = 0.5)')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate (Rasio Alarm Palsu)')
plt.ylabel('True Positive Rate (Recall / Sensitivitas)')
plt.title('Receiver Operating Characteristic (ROC) Curve', fontsize=16)
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
print(f"\nNilai Area Under Curve (AUC): {auc_score:.4f}")
print("Semakin nilai AUC mendekati 1.0, semakin baik kemampuan model dalam membedakan kedua kelas.")

print("\n d. Analisis Kinerja pada Berbagai Ambang Batas Klasifikasi:")
# Ini lebih relevan untuk klasifikasi biner dengan output sigmoid
# Karena kita menggunakan softmax, y_pred_probs sudah merupakan probabilitas kelas.
# Namun, kita masih bisa mengevaluasi pengaruh ambang batas pada probabilitas kelas 'precursor'
thresholds_range = np.arange(0.1, 1.0, 0.05)
best_f1_prec = -1
best_threshold = 0.5
performance_data = []

print("{:<10} {:<10} {:<10} {:<10} {:<10}".format("Threshold", "P_Prec", "R_Prec", "F1_Prec", "Accuracy"))
for threshold in thresholds_range:
    # Mengklasifikasikan sebagai 'precursor' jika probabilitas 'precursor' > threshold
    y_pred_tuned = (y_score > threshold).astype(int)
    
    try:
        # Mengatasi kasus di mana semua prediksi jatuh ke satu kelas
        unique_preds = np.unique(y_pred_tuned)
        if len(unique_preds) == 1:
            if unique_preds[0] == 0: # Semua diprediksi Normal
                p_prec = 0.0; r_prec = 0.0; f1_prec = 0.0
                accuracy_tuned = np.mean(y_true == y_pred_tuned)
            else: # Semua diprediksi Prekursor
                tp = np.sum((y_true == 1) & (y_pred_tuned == 1))
                fp = np.sum((y_true == 0) & (y_pred_tuned == 1))
                fn = np.sum((y_true == 1) & (y_pred_tuned == 0))
                p_prec = tp / (tp + fp + 1e-10) # Tambah epsilon untuk menghindari DivByZero
                r_prec = tp / (tp + fn + 1e-10)
                f1_prec = 2 * (p_prec * r_prec) / (p_prec + r_prec + 1e-10) if (p_prec + r_prec) > 0 else 0.0
                accuracy_tuned = np.mean(y_true == y_pred_tuned)
        else:
            # Gunakan classification_report untuk metrik yang lebih andal
            cr = classification_report(y_true, y_pred_tuned, output_dict=True, zero_division=0, labels=[0, 1])
            p_prec = cr['1']['precision'] # Ambil metrik untuk kelas 'precursor' (label 1)
            r_prec = cr['1']['recall']
            f1_prec = cr['1']['f1-score']
            accuracy_tuned = cr['accuracy']
    except KeyError:
        p_prec = 0.0; r_prec = 0.0; f1_prec = 0.0
        accuracy_tuned = np.mean(y_true == y_pred_tuned)

    performance_data.append({'Threshold': threshold, 'P_Prec': p_prec, 'R_Prec': r_prec, 'F1_Prec': f1_prec, 'Accuracy': accuracy_tuned})
    print("{:<10.2f} {:<10.2f} {:<10.2f} {:<10.2f} {:<10.2f}".format(threshold, p_prec, r_prec, f1_prec, accuracy_tuned))
    
    if f1_prec > best_f1_prec:
        best_f1_prec = f1_prec
        best_threshold = threshold

print(f"\nAmbang batas terbaik berdasarkan F1-score Prekursor: {best_threshold:.2f} (F1 = {best_f1_prec:.2f})")

df_performance = pd.DataFrame(performance_data)

plt.figure(figsize=(12, 6))
plt.plot(df_performance['Threshold'], df_performance['P_Prec'], label='Precision Prekursor', marker='o')
plt.plot(df_performance['Threshold'], df_performance['R_Prec'], label='Recall Prekursor', marker='o')
plt.plot(df_performance['Threshold'], df_performance['F1_Prec'], label='F1-Score Prekursor', marker='o', linestyle='--')

plt.title('Precision, Recall, dan F1-Score Prekursor vs. Threshold', fontsize=16)
plt.xlabel('Ambang Batas (Threshold)')
plt.ylabel('Skor')
plt.axvline(x=0.5, color='gray', linestyle=':', label='Threshold Default (0.5)')
plt.axvline(x=best_threshold, color='green', linestyle='--', label=f'Best F1 Threshold ({best_threshold:.2f})')
plt.legend()
plt.grid(True)
plt.show()

print("Sel 13 Selesai.")
print(f"\n{'='*25} LAPORAN SELESAI {'='*25}")
```
